```python
'''

Connect to the Presentation:
Rickish Partitions​ : Silly Ways a Sum of Some Things Become Whole​

Here, we run through an example of how to generate a Rickish Partition for 7.
'''
```

```python
## VARIABLES AND EMPTY LISTS


a = 1000
board_history = [] ## lake of all paths which we are to experience via Mancala generation
partition_paths = []
```

```python
## RUN RICKISH PARTITION PROCESS FOR 7


for b in range(2, a): ## for each number less than the number
    
    ## "forcible doubling up"
    ## Since the Mancala process deterministically divides an even number by 2, we take a number and make it as big as allowable ...
    ## ... (i.e. still smaller than a) before beginning the process. This avoids duplication of efforts and replication of paths ...
    ## ... down the road.
    print("The Mancala process is beginning for (", b, ",", a - b, ").") 
    while 2 * b < a:
        b = b * 2
    board = [b, a - b] ## create the board
    active_position = 0 ## initial active position
    inactive_position = 1 ## initial inactive position
    bin_continue = 1 ## binary switch to continue
    turn_history = []
    ## if we have at any point seen this number in a Mancala path, skipadelphia!
    if b in board_history:
        bin_continue = 0
    while bin_continue == 1: ## continue until there is a reason not to
        board_history.append(board[active_position]) ## add to board history
        turn_history.append(board[active_position]) ## add to board history
        active_position_value = board[active_position] ## save the value
        inactive_position_value = board[inactive_position] ## save the value
        ## if active position is even, reduce
        if board[active_position] % 2 == 0: 
            board[active_position] = active_position_value/2
            board[inactive_position] = inactive_position_value + active_position_value/2
        ## distribute and change active position
        else:
            ## distribute
            board[active_position] = (active_position_value - 1)/2
            board[inactive_position] = inactive_position_value + (active_position_value + 1)/2
            ## change active position
            if active_position == 1:
                active_position = 0
                inactive_position = 1
            else:
                active_position = 1
                inactive_position = 0
        ## if we've seen the new result before, flip the switch
        if board[active_position] in turn_history:
            bin_continue = 0
        ## if we've converged, or the active position value is 1 ...
        ## ... (in which case we will converge, and the code is not otherwise written to handle 1) ...
        ## then flip the switch, append data to lists as necessary
        if board[active_position] == a or board[inactive_position] == a or board[active_position] < 2 :
            bin_continue = 0
            turn_history.append(1)
            turn_history.append(a)
            board_history.append(1)
            board_history.append(a)
    partition_paths.append(turn_history)
```

```python
## CLEAN UP AND SORT PARTITION PATHS


## The above process yields empty sets, so I go ahead and get rid of them.
partition_paths_pre = [x for x in partition_paths if x]
## Then I sort by length; this will be instrumental in cleaning up duplicates in the next step
partition_paths_pre = sorted(partition_paths_pre, key=len, reverse = True)

## now to get rid of partition paths that exist within other partition paths
## for each partition path, if we find that it existed in a previously accepted partition path ...
## ... then do not include it
partition_paths = []
for d, b in enumerate(partition_paths_pre):
    keep_this_thing = 1
    ## first partition path (the longest one) gets accepted by default
    if d == 0:
        partition_paths.append(b)
    else:
        ## if we find one of the elements in the path existed in other path, then we revoke inclusion
        for c in range(0, d):
            if b[0] in partition_paths_pre[c]:
                keep_this_thing = keep_this_thing * 0
        ## if it survives all rounds of inspection, then add it to the list of partition paths
        if keep_this_thing == 1:
            partition_paths.append(b)
```

```python
## PRINT THE MANCALA PATHS


print('The following are the unique Mancala paths for', a, ':')
print('---------------------------------------------------------------------------------------------------------------------------------')
print('\n')
for i in partition_paths:
    does_path_converge = 0
    print_string = ''
    for ii in i:
        if ii == a:
            converge_or_diverge = 'converges'
        else:
            converge_or_diverge = 'diverges'
        print_string = print_string + str(ii) + ' --> '
    print_string = print_string + converge_or_diverge
    print('This path is', len(i), 'element(s) long.')
    print(print_string)
    print('\n')
```

```python
## GET THE RICKISH PARTITIONS


rickish_partition = []
for b in partition_paths:
    if b[-1] == a:
        rickish_partition.append([1,len(b)])
    else:
        rickish_partition.append([0,len(b)])
```

```python
## PRINT THE RICKISH PARTITION


print('The Rickish partition for', a, 'is:')
print_string = str(a) + ' = '
for i in rickish_partition:
    print_string = print_string + str(i[1]) + ' + '
print_string = print_string[:-3]
print(print_string)
```
